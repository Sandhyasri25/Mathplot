```python
import matplotlib.pyplot as plt
import numpy as np
class_A = np.random.normal(75, 10, 100) 
class_B = np.random.normal(70, 15, 100)  
class_C = np.random.normal(80, 5, 100)   
data = [class_A, class_B, class_C]
plt.figure(figsize=(8, 6))
plt.violinplot(data, showmeans=True, showmedians=True)
plt.title('Distribution of Student Scores Across Classes')
plt.xticks([1, 2, 3], ['Class A', 'Class B', 'Class C'])
plt.ylabel('Scores')
plt.grid(True)
plt.show()
```
